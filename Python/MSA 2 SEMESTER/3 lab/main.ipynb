# Завдання 12. Дискримінантний аналіз

**Варіант 4:** Ознаки 1, 6, 8, 9.
За результатами попереднього розбиття (методом $k$-середніх) маємо два класи:
* **Клас 1:** ПАР, Південна Корея, Північна Корея, Польща, Португалія, Росія, Саудівська Аравія.
* **Клас 2:** Нова Зеландія, Норвегія, ОАЕ.

**Нові об'єкти для класифікації:**
1. Сінгапур (41)
2. США (42)
3. Таїланд (43)

**Алгоритм:** Побудова канонічної дискримінантної функції Фішера (Алгоритм 5.1) за допомогою об'єднаної коваріаційної матриці. Розрахунки проводяться на вихідних (нестандартизованих) даних, оскільки коваріаційна матриця автоматично враховує масштаб ознак.

```python
import numpy as np

np.set_printoptions(precision=4, suppress=True)

# Дані Класу 1 (7 об'єктів)
X1 = np.array([
    [43900, 3128,  49, 76],  # ПАР
    [45000, 6627,  72, 96],  # Південна Корея
    [23100, 1000,  60, 99],  # Північна Корея
    [38600, 4429,  62, 99],  # Польща
    [10500, 9000,  34, 85],  # Португалія
    [149200, 6680, 74, 99],  # Росія
    [18000, 6651,  77, 62]   # Саудівська Аравія
], dtype=float)

# Дані Класу 2 (3 об'єкти)
X2 = np.array([
    [3524, 14381, 84, 99],   # Нова Зеландія
    [4300, 17755, 75, 99],   # Норвегія
    [2800, 14193, 81, 68]    # ОАЕ
], dtype=float)

# Нові об'єкти для перевірки
X_new = np.array([
    [2900, 14990, 100, 88],  # 41 Сінгапур
    [260800, 23474, 75, 97], # 42 США
    [59400, 1800, 22, 93]    # 43 Таїланд
], dtype=float)

new_names = ["Сінгапур", "США", "Таїланд"]

print(f"Кількість об'єктів у Класі 1: m1 = {len(X1)}")
print(f"Кількість об'єктів у Класі 2: m2 = {len(X2)}")
```

```python
# Крок 1: Знаходимо вектори середніх значень
mean1 = np.mean(X1, axis=0)
mean2 = np.mean(X2, axis=0)
print("=== Крок 1: Вектори середніх значень ===")
print(f"Клас 1 (mean1): {mean1}")
print(f"Клас 2 (mean2): {mean2}\n")

# Крок 2: Матриці центрованих значень
X1_centered = X1 - mean1
X2_centered = X2 - mean2

# Крок 3: Розраховуємо об'єднану коваріаційну матрицю S
m1, m2 = len(X1), len(X2)
# Формула 5.2: S = 1 / (m1 + m2 - 2) * (X1~^T * X1~ + X2~^T * X2~)
S = (X1_centered.T @ X1_centered + X2_centered.T @ X2_centered) / (m1 + m2 - 2)
print("=== Крок 3: Об'єднана коваріаційна матриця S ===")
print(S)
print()

# Крок 4: Обчислюємо обернену матрицю S^(-1)
S_inv = np.linalg.inv(S)
print("=== Крок 4: Обернена коваріаційна матриця S^(-1) ===")
print(S_inv)
print()

# Крок 5: Коефіцієнти дискримінантної функції A = S^(-1) * (mean1 - mean2)
# Примітка: за формулою 5.3 A = S^(-1) * (X_1 - X_2), де X_1, X_2 - вектори середніх
A = S_inv @ (mean1 - mean2)
print("=== Крок 5: Вектор коефіцієнтів дискримінантної функції (A) ===")
print(f"A = {A}\n")

# Крок 6: Середні значення функції та константа дискримінації C
f1_mean = mean1 @ A
f2_mean = mean2 @ A
C = (f1_mean + f2_mean) / 2

print("=== Крок 6: Середні значення функції та константа C ===")
print(f"Середнє для Класу 1 (f1_mean) = {f1_mean:.4f}")
print(f"Середнє для Класу 2 (f2_mean) = {f2_mean:.4f}")
print(f"Константа дискримінації (C) = {C:.4f}\n")

# Визначаємо правило: який клас має більше значення функції
larger_class = 1 if f1_mean > f2_mean else 2
smaller_class = 2 if f1_mean > f2_mean else 1

print(f"Правило класифікації: якщо f(x) > {C:.4f}, відносимо до Класу {larger_class}, інакше - до Класу {smaller_class}.\n")

# Крок 7-8: Класифікація нових об'єктів
print("=== Крок 7-8: Класифікація нових об'єктів ===")
for i, x_k in enumerate(X_new):
    f_k = x_k @ A
    if f_k > C:
        assigned_class = larger_class
    else:
        assigned_class = smaller_class

    print(f"{new_names[i]}:")
    print(f"  Ознаки: {x_k}")
    print(f"  Значення функції f_k = {f_k:.4f}")
    print(f"  -> Віднесено до Класу {assigned_class}\n")
```
