# Завдання 12. Дискримінантний аналіз (Оновлений список країн)

**Варіант 9:** Ознаки 6, 7, 8, 9.
За результатами розбиття методом $k$-середніх маємо:
* **Клас 1:** Гонконг (1 об'єкт - виражений математичний "викид").
* **Клас 2:** Ефіопія, Єгипет, Замбія, Індія, Ірландія, Іспанія, Італія, Канада, Китай (9 об'єктів).

**Нові об'єкти для класифікації (6 країн):**
1. Бангладеш (4)
2. Буркіна-Фасо (8)
3. В'єтнам (10)
4. Гондурас (12)
5. Філіппіни (46)
6. Чилі (49)

**Алгоритм:** Розрахунок лінійної дискримінантної функції (ЛДФ) за Алгоритмом 5.1 на нестандартизованих даних. Внутрішньогрупова варіація для Класу 1 дорівнює нулю, тому матриця $\boldsymbol{S}$ визначатиметься виключно дисперсією Класу 2.

```python
import numpy as np

np.set_printoptions(precision=4, suppress=True)

# Клас 1 (1 об'єкт - Гонконг)
X1 = np.array([
    [14641, 5494, 94, 77]
], dtype=float)

# Клас 2 (9 об'єктів)
X2 = np.array([
    [122, 47, 12, 24],       # Ефіопія
    [748, 57, 44, 48],       # Єгипет
    [573, 11, 42, 73],       # Замбія
    [275, 283, 26, 52],      # Індія
    [12170, 51, 57, 98],     # Ірландія
    [13047, 77, 78, 95],     # Іспанія
    [17500, 188, 69, 97],    # Італія
    [19904, 2.8, 77, 97],    # Канада
    [377, 124, 26, 78]       # Китай
], dtype=float)

# 6 нових об'єктів для перевірки
X_new = np.array([
    [202,  800, 16, 35],     # 4 Бангладеш
    [357,  36,  15, 18],     # 8 Буркіна-Фасо
    [230,  218, 20, 88],     # 10 В'єтнам
    [1030, 46,  44, 73],     # 12 Гондурас
    [867,  221, 43, 90],     # 46 Філіппіни
    [2591, 18,  85, 93]      # 49 Чилі
], dtype=float)

new_names = [
    "Бангладеш", "Буркіна-Фасо", "В'єтнам",
    "Гондурас", "Філіппіни", "Чилі"
]

print(f"Кількість об'єктів у Класі 1: m1 = {len(X1)}")
print(f"Кількість об'єктів у Класі 2: m2 = {len(X2)}")
```

```python
# Крок 1: Вектори середніх значень
mean1 = np.mean(X1, axis=0)
mean2 = np.mean(X2, axis=0)
print("=== Крок 1: Вектори середніх значень ===")
print(f"Клас 1 (mean1): {mean1}")
print(f"Клас 2 (mean2): {mean2}\n")

# Крок 2: Матриці центрованих значень
X1_centered = X1 - mean1
X2_centered = X2 - mean2

# Крок 3: Об'єднана коваріаційна матриця S
m1, m2 = len(X1), len(X2)
S = (X1_centered.T @ X1_centered + X2_centered.T @ X2_centered) / (m1 + m2 - 2)
print("=== Крок 3: Об'єднана коваріаційна матриця S ===")
print(S)
print()

# Крок 4: Обернена коваріаційна матриця S^(-1)
S_inv = np.linalg.inv(S)
print("=== Крок 4: Обернена матриця S^(-1) ===")
print(S_inv)
print()

# Крок 5: Вектор коефіцієнтів дискримінантної функції (A)
A = S_inv @ (mean1 - mean2)
print("=== Крок 5: Вектор коефіцієнтів (A) ===")
print(f"A = {A}\n")

# Крок 6: Константа дискримінації C
f1_mean = mean1 @ A
f2_mean = mean2 @ A
C = (f1_mean + f2_mean) / 2

print("=== Крок 6: Константа дискримінації ===")
print(f"Середнє f(x) для Класу 1 = {f1_mean:.4f}")
print(f"Середнє f(x) для Класу 2 = {f2_mean:.4f}")
print(f"Константа C = {C:.4f}\n")

larger_class = 1 if f1_mean > f2_mean else 2
smaller_class = 2 if f1_mean > f2_mean else 1
print(f"Правило класифікації: якщо f(x) > {C:.4f} -> відносимо до Класу {larger_class}, інакше -> до Класу {smaller_class}\n")

# Крок 7-8: Класифікація нових країн
print("=== Крок 7-8: Результати класифікації нових об'єктів ===")
for i, x_k in enumerate(X_new):
    f_k = x_k @ A
    assigned_class = larger_class if f_k > C else smaller_class

    print(f"{new_names[i]:<15} | f_k = {f_k:>10.4f} | -> Клас {assigned_class}")
```
