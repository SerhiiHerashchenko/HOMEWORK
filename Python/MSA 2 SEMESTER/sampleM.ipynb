```python
import pandas as pd
import math
from IPython.display import display
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram
import numpy as np

# ==========================================
# ДАНІ ВАРІАНТА 4
# ==========================================
data_var9 = {
    'Країна': ['Гонконг', 'Ефіопія', 'Єгипет', 'Замбія', 'Індія',
               'Ірландія', 'Іспанія', 'Італія', 'Канада', 'Китай'],
    'Ознака_6': [14641, 122, 748, 573, 275, 12170, 13047, 17500, 19904, 377],
    'Ознака_7': [5494, 47, 57, 11, 283, 51, 77, 188, 2.8, 124], # У Канади 2,8
    'Ознака_8': [94, 12, 44, 42, 26, 57, 78, 69, 77, 26],
    'Ознака_9': [77, 24, 48, 73, 52, 98, 95, 97, 97, 78]
}
df = pd.DataFrame(data_var9).set_index('Країна')

print("1. Вихідні дані:")
display(df)

# --- РУЧНА НОРМАЛІЗАЦІЯ ---
df_scaled = pd.DataFrame(index=df.index, columns=df.columns)
for col in df.columns:
    col_mean = sum(df[col]) / len(df[col])
    # Розрахунок стандартного відхилення
    variance = sum((x - col_mean) ** 2 for x in df[col]) / len(df[col])
    col_std = math.sqrt(variance)
    # Нормалізація значень
    df_scaled[col] = [(x - col_mean) / col_std for x in df[col]]

print("\n2. Нормалізовані дані (вручну):")
display(df_scaled.round(4))

# --- РУЧНИЙ РОЗРАХУНОК МАТРИЦІ ЕВКЛІДОВИХ ВІДСТАНЕЙ ---
n = len(df_scaled)
dist_matrix = pd.DataFrame(0.0, index=df.index, columns=df.index)

for i in range(n):
    for j in range(n):
        if i != j:
            # Сума квадратів різниць по кожній ознаці
            sum_sq = sum((df_scaled.iloc[i, k] - df_scaled.iloc[j, k])**2 for k in range(len(df_scaled.columns)))
            dist_matrix.iloc[i, j] = math.sqrt(sum_sq)

print("\n3. Початкова матриця відстаней D0 (вручну):")
initial_dist_matrix = dist_matrix.copy() # Збережемо для дивізимного методу
display(dist_matrix.round(3))
```

```python
print("=== АГЛОМЕРАТИВНИЙ МЕТОД (КРОК ЗА КРОКОМ) ===")

current_dist = initial_dist_matrix.copy()
n_samples = len(current_dist)

# Для побудови дендрограми нам потрібно вручну зібрати матрицю Z (linkage matrix)
# Формат кожного рядка Z: [id_кластера_1, id_кластера_2, відстань, кількість_елементів]
Z = np.zeros((n_samples - 1, 4))

# Зіставляємо кожному початковому об'єкту унікальний ID від 0 до n-1
cluster_ids = {name: i for i, name in enumerate(current_dist.index)}
# Зберігаємо кількість об'єктів у кожному кластері
cluster_sizes = {name: 1 for name in current_dist.index}

next_cluster_id = n_samples
step = 0

while len(current_dist) > 1:
    # 1. Шукаємо мінімальну відстань (не 0) у матриці
    min_dist = float('inf')
    merge_a, merge_b = None, None

    for i in range(len(current_dist.index)):
        for j in range(i + 1, len(current_dist.columns)):
            d = current_dist.iloc[i, j]
            if d < min_dist:
                min_dist = d
                merge_a = current_dist.index[i]
                merge_b = current_dist.columns[j]

    # Формуємо нову назву об'єднаного кластера для виводу
    new_cluster_name = f"({merge_a.split('_')[0]},{merge_b.split('_')[0]})"

    print(f"\nКрок {step + 1}: Мінімальна відстань d = {min_dist:.3f}. Об'єднуємо '{merge_a}' та '{merge_b}'.")

    # --- Записуємо дані у нашу ручну матрицю Z для дендрограми ---
    id_a = cluster_ids[merge_a]
    id_b = cluster_ids[merge_b]
    new_size = cluster_sizes[merge_a] + cluster_sizes[merge_b]

    Z[step, 0] = id_a
    Z[step, 1] = id_b
    Z[step, 2] = min_dist
    Z[step, 3] = new_size

    # Реєструємо новий кластер
    cluster_ids[new_cluster_name] = next_cluster_id
    cluster_sizes[new_cluster_name] = new_size
    next_cluster_id += 1
    # --------------------------------------------------------------

    # 2. Створюємо нову матрицю відстаней
    new_cols = [c for c in current_dist.columns if c not in [merge_a, merge_b]]
    new_dist = pd.DataFrame(0.0, index=new_cols + [new_cluster_name], columns=new_cols + [new_cluster_name])

    # Переносимо старі відстані для кластерів, які не змінилися
    for c1 in new_cols:
        for c2 in new_cols:
            new_dist.loc[c1, c2] = current_dist.loc[c1, c2]

    # 3. Рахуємо відстані для нового кластера (метод "найближчого сусіда")
    for c in new_cols:
        dist_to_new = min(current_dist.loc[merge_a, c], current_dist.loc[merge_b, c])
        new_dist.loc[new_cluster_name, c] = dist_to_new
        new_dist.loc[c, new_cluster_name] = dist_to_new

    current_dist = new_dist
    display(current_dist.round(3))
    step += 1

# --- ПОБУДОВА ДЕНДРОГРАМИ ---
print("\nБУДУЄМО ДЕНДРОГРАМУ НА ОСНОВІ РУЧНИХ ОБЧИСЛЕНЬ:")
plt.figure(figsize=(10, 6))
plt.title('Дендрограма (Агломеративний метод "ближнього сусіда")')

# Використовуємо dendrogram виключно для малювання нашої зібраної вручну матриці Z
dendrogram(
    Z,
    labels=[name.split('_')[1] if '_' in name else name for name in initial_dist_matrix.index],
    leaf_rotation=45,
    leaf_font_size=11,
    color_threshold=0 # Щоб усі лінії були одного кольору, як у методичці
)

plt.ylabel('Відстань об\'єднання (d)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
print("=== ДИВІЗИМНИЙ МЕТОД (КРОК ЗА КРОКОМ) ===")

# Початковий стан: 1 кластер, що містить усі країни
clusters = [list(initial_dist_matrix.index)]
step = 1

while len(clusters) < len(initial_dist_matrix):
    max_internal_dist = -1
    cluster_to_split = None
    center1 = None
    center2 = None

    # 1. Знаходимо кластер з найбільшою внутрішньою відстанню
    for cluster in clusters:
        if len(cluster) > 1:
            # Шукаємо максимальну відстань між елементами всередині цього кластера
            for i in range(len(cluster)):
                for j in range(i + 1, len(cluster)):
                    obj1, obj2 = cluster[i], cluster[j]
                    d = initial_dist_matrix.loc[obj1, obj2]
                    if d > max_internal_dist:
                        max_internal_dist = d
                        cluster_to_split = cluster
                        center1 = obj1
                        center2 = obj2

    if cluster_to_split is None:
        break

    # 2. Формуємо два нові класи на базі знайдених центрів
    new_class_1 = [center1]
    new_class_2 = [center2]

    print(f"\nКрок {step}: Розглядаємо кластер розміром {len(cluster_to_split)} об'єктів.")
    print(f"Найбільша відстань Δd = {max_internal_dist:.3f} між '{center1}' та '{center2}'.")
    print("Вони стають центрами розбиття. Розподіляємо інші об'єкти:")

    # 3. Розподіляємо інші об'єкти кластера
    for obj in cluster_to_split:
        if obj != center1 and obj != center2:
            d_to_c1 = initial_dist_matrix.loc[obj, center1]
            d_to_c2 = initial_dist_matrix.loc[obj, center2]

            if d_to_c1 < d_to_c2:
                new_class_1.append(obj)
                print(f"  - '{obj}': до {center1} ({d_to_c1:.3f}) < до {center2} ({d_to_c2:.3f}) -> йде у Клас 1")
            else:
                new_class_2.append(obj)
                print(f"  - '{obj}': до {center1} ({d_to_c1:.3f}) > до {center2} ({d_to_c2:.3f}) -> йде у Клас 2")

    # Оновлюємо список кластерів
    clusters.remove(cluster_to_split)
    clusters.append(new_class_1)
    clusters.append(new_class_2)

    print(f"\nРЕЗУЛЬТАТ РОЗБИТТЯ НА КРОЦІ {step}:")
    print(f"Клас 1: {[c.split('_')[0] for c in new_class_1]}")
    print(f"Клас 2: {[c.split('_')[0] for c in new_class_2]}")
    step += 1
```
