```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx

# Дані 9 варіанту
data = np.array([
    [17800,  16848,   2.3, 100, 1.38, 74, 80],
    [10300,   6500,  50.0,  99, 0.32, 66, 76],
    [156600,  2354,  18.0,  81, 1.28, 57, 67],
    [58400,  15974, 237.0,  99, 0.20, 74, 80],
    [73100,    230, 218.0,  88, 1.78, 63, 68],
    [6500,     383, 231.0,  53, 1.63, 43, 47],
    [5600,    1030,  46.0,  73, 2.73, 65, 70]
], dtype=float)

# Назви країн згідно з рядками
feature_names = ['Австралія', 'Білорусь', 'Бразилія', 'Велика Британія', "В'єтнам", 'Гаїті', 'Гондурас']
n_features = len(feature_names)

print("Розмірність вхідних даних:", data.shape)
print("Країни для аналізу:", feature_names)
```

```python
# 1. Стандартизація даних по стовпцях (ознаках), щоб зрівняти масштаби
means = np.mean(data, axis=0)
stds = np.std(data, axis=0)
data_std = (data - means) / stds

print("--- Стандартизовані дані ---")
print(np.round(data_std, 2))

# 2. Відстані між КРАЇНАМИ (РЯДКАМИ)
dist_matrix = np.zeros((n_features, n_features))
for i in range(n_features):
    for j in range(n_features):
        # Зверни увагу: рахуємо відстань між i-тим та j-тим РЯДКОМ (data_std[i, :])
        dist_matrix[i, j] = np.sqrt(np.sum((data_std[i, :] - data_std[j, :])**2))

print("\n--- Матриця відстаней між країнами (D) ---")
print(np.round(dist_matrix, 2))

# 3. Кореляції між КРАЇНАМИ (на основі стандартизованих даних)
# Знаходимо середнє та дисперсію для кожного рядка (країни)
means_rows = np.mean(data_std, axis=1)
stds_rows = np.std(data_std, axis=1)

corr_matrix = np.zeros((n_features, n_features))
for i in range(n_features):
    for j in range(n_features):
        cov = np.mean((data_std[i, :] - means_rows[i]) * (data_std[j, :] - means_rows[j]))
        # Уникаємо ділення на нуль
        if stds_rows[i] * stds_rows[j] != 0:
            corr_matrix[i, j] = cov / (stds_rows[i] * stds_rows[j])
        else:
            corr_matrix[i, j] = 0

print("\n--- Матриця кореляцій між країнами (R) ---")
print(np.round(corr_matrix, 3))
```

```python
print("=== МЕТОД ДЕНДРИТІВ: Побудова ===")
edges = []

# 1. Зв'язки першого порядку
for i in range(n_features):
    min_dist = float('inf')
    min_j = -1
    for j in range(n_features):
        if i != j and dist_matrix[i, j] < min_dist:
            min_dist = dist_matrix[i, j]
            min_j = j

    # Зберігаємо ребро, відсортувавши індекси вершин, щоб уникнути дублікатів (i-j та j-i)
    edge = (min(i, min_j), max(i, min_j), min_dist)
    if edge not in edges:
        edges.append(edge)

print(f"Зв'язки 1-го порядку:")
for u, v, w in edges:
    print(f" {feature_names[u]} - {feature_names[v]} (d = {w:.3f})")

# Функція для знаходження ізольованих скупчень (компонент зв'язності)
def get_components(edges_list, n):
    components = []
    visited = set()

    def dfs(node, current_comp):
        visited.add(node)
        current_comp.add(node)
        for u, v, w in edges_list:
            neighbor = v if u == node else u if v == node else None
            if neighbor is not None and neighbor not in visited:
                dfs(neighbor, current_comp)

    for i in range(n):
        if i not in visited:
            comp = set()
            dfs(i, comp)
            components.append(comp)

    return components

# 2. Зв'язки вищих порядків
order = 2
while True:
    components = get_components(edges, n_features)

    if len(components) == 1:
        break # Всі вершини з'єднані

    print(f"\nПоточні скупчення: {[{feature_names[idx] for idx in c} for c in components]}")

    # Шукаємо найкоротший зв'язок між РІЗНИМИ скупченнями
    min_dist = float('inf')
    best_edge = None

    for i in range(len(components)):
        for j in range(i + 1, len(components)):
            for u in components[i]:
                for v in components[j]:
                    if dist_matrix[u, v] < min_dist:
                        min_dist = dist_matrix[u, v]
                        best_edge = (min(u, v), max(u, v), min_dist)

    edges.append(best_edge)
    print(f"Додано зв'язок {order}-го порядку: {feature_names[best_edge[0]]} - {feature_names[best_edge[1]]} (d = {best_edge[2]:.3f})")
    order += 1

print(f"\nДендрит побудовано. Всього зв'язків: {len(edges)}")
```

```python
print("=== МЕТОД ДЕНДРИТІВ: Розбиття на класи ===")

# Сортуємо зв'язки за спаданням їхньої довжини
sorted_edges = sorted(edges, key=lambda x: x[2], reverse=True)
d_values = [e[2] for e in sorted_edges]

print("Відстані d_i за спаданням:")
for i, d in enumerate(d_values):
    print(f" d_{i+1} = {d:.3f}")

# Розрахунок відношень сусідніх зв'язків i_k = d_{k-1}/d_k
i_values = []
for k in range(len(d_values) - 1):
    i_values.append(d_values[k] / d_values[k+1])

print("\nВідношення довжин сусідніх зв'язків (i_k):")
for k, i_val in enumerate(i_values):
    print(f" i_{k+2} = {i_val:.3f}")

# Шукаємо, де виконується i_k < i_{k+1}
candidates = []
for k_idx in range(len(i_values) - 1):
    if i_values[k_idx] < i_values[k_idx+1]:
        k_real = k_idx + 2
        candidates.append((k_real, i_values[k_idx]))

if candidates:
    print(f"\nУмова i_k < i_{k+1} виконується для: {[(f'k={c[0]}', round(c[1], 3)) for c in candidates]}")
    # Вибираємо те k, де i_k мінімальне
    best_k = min(candidates, key=lambda x: x[1])[0]
    print(f"Найкраще розбиття при k = {best_k}.")
else:
    print("\nЯвного виконання умови немає. Розбиваємо за найбільшим стрибком.")
    best_k = np.argmax(i_values) + 2

edges_to_remove = best_k - 1
print(f"Видаляємо {edges_to_remove} найдовших зв'язків з дендрита.")

final_edges = sorted_edges[edges_to_remove:]
final_classes = get_components(final_edges, n_features)

print("\nКЛАСИ ОЗНАК (Метод Дендритів):")
for idx, comp in enumerate(final_classes):
    class_names = [feature_names[i] for i in comp]
    print(f" Клас {idx+1}: {class_names}")
```

```python
print("=== МЕТОД КОРЕЛЯЦІЙНИХ ПЛЕЯД ===")

# Зануляємо діагональ, щоб не брати її до уваги
np.fill_diagonal(corr_matrix, 0)
abs_corr = np.abs(corr_matrix)

pleiades_edges = []
in_graph = set()

# Крок 1: Глобальний максимум кореляції
max_r = -1
u_start, v_start = -1, -1
for i in range(n_features):
    for j in range(i+1, n_features):
        if abs_corr[i, j] > max_r:
            max_r = abs_corr[i, j]
            u_start, v_start = i, j

pleiades_edges.append((u_start, v_start, corr_matrix[u_start, v_start]))
in_graph.add(u_start)
in_graph.add(v_start)
print(f"Крок 1: Перша пара: {feature_names[u_start]} - {feature_names[v_start]} (|r| = {max_r:.3f})")

# Крок 2: Приєднання вершин
for step in range(2, n_features):
    max_r = -1
    best_u, best_v = -1, -1

    # Шукаємо максимальну кореляцію між множиною В графі та ПОЗА графом
    for u in in_graph:
        for v in range(n_features):
            if v not in in_graph:
                if abs_corr[u, v] > max_r:
                    max_r = abs_corr[u, v]
                    best_u, best_v = u, v

    pleiades_edges.append((best_u, best_v, corr_matrix[best_u, best_v]))
    in_graph.add(best_v)
    print(f"Крок {step}: Приєднуємо {feature_names[best_v]} до {feature_names[best_u]} (|r| = {max_r:.3f})")

# Крок 3: Порогове значення. Візьмемо r0 = 0.65 як межу помітної кореляції.
# (Можеш змінити це значення на свій розсуд для дослідження)
threshold_r = 0.65
print(f"\nЗадаємо поріг r = {threshold_r}. Видаляємо всі зв'язки, де |r| < {threshold_r}.")

final_pleiades_edges = [e for e in pleiades_edges if abs(e[2]) >= threshold_r]
pleiades_classes = get_components(final_pleiades_edges, n_features)

print("\nПЛЕЯДИ ОЗНАК:")
for idx, comp in enumerate(pleiades_classes):
    class_names = [feature_names[i] for i in comp]
    print(f" Плеяда {idx+1}: {class_names}")
```

```python
plt.figure(figsize=(16, 6))

# --- Граф Дендрита ---
plt.subplot(1, 2, 1)
G_dendrite = nx.Graph()
for i in range(n_features):
    G_dendrite.add_node(feature_names[i])
for u, v, w in final_edges:
    G_dendrite.add_edge(feature_names[u], feature_names[v], weight=round(w, 2))

color_map_d = []
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0', '#ffb3e6']
for node in G_dendrite.nodes():
    for idx, comp in enumerate(final_classes):
        if list(G_dendrite.nodes()).index(node) in comp:
            color_map_d.append(colors[idx % len(colors)])

pos_d = nx.spring_layout(G_dendrite, seed=15)
nx.draw(G_dendrite, pos_d, with_labels=True, node_color=color_map_d,
        node_size=2500, font_size=9, font_weight='bold', edge_color='gray')
labels_d = nx.get_edge_attributes(G_dendrite, 'weight')
nx.draw_networkx_edge_labels(G_dendrite, pos_d, edge_labels=labels_d)
plt.title('Розбиття методом Дендритів (відстані)')

# --- Граф Плеяд ---
plt.subplot(1, 2, 2)
G_pleiades = nx.Graph()
for i in range(n_features):
    G_pleiades.add_node(feature_names[i])
for u, v, w in final_pleiades_edges:
    G_pleiades.add_edge(feature_names[u], feature_names[v], weight=round(w, 2))

color_map_p = []
for node in G_pleiades.nodes():
    for idx, comp in enumerate(pleiades_classes):
        if list(G_pleiades.nodes()).index(node) in comp:
            color_map_p.append(colors[idx % len(colors)])

pos_p = nx.spring_layout(G_pleiades, seed=15)
nx.draw(G_pleiades, pos_p, with_labels=True, node_color=color_map_p,
        node_size=2500, font_size=9, font_weight='bold', edge_color='red')
labels_p = nx.get_edge_attributes(G_pleiades, 'weight')
nx.draw_networkx_edge_labels(G_pleiades, pos_p, edge_labels=labels_p)
plt.title(f'Кореляційні Плеяди (поріг |r| >= {threshold_r})')

plt.tight_layout()
plt.show()
```
