**Мета роботи:** Провести аналіз ознак методом головних компонент (МГК) без використання готових бібліотек машинного навчання, реалізуючи алгоритм за математичними формулами.

### Ініціалізація даних та імпорт бібліотек
Задамо початкову матрицю $X$, де рядки — це об'єкти (1-10), а стовпці — показники підприємств (ознаки 7, 11, 12, 13, 17 з таблиці А.6).

```python
import numpy as np
import pandas as pd
from scipy.stats import chi2 # Використовуємо лише для отримання табличних критичних значень

# Початкові дані (Варіант 4: ознаки 7, 11, 12, 13, 17)
# Рядки - об'єкти (1-10), стовпці - ознаки
data = [
    [1.37, 26006, 167.69, 47750, 17.72],
    [1.49, 23935, 186.10, 50391, 18.39],
    [1.44, 22589, 220.45, 43149, 26.46],
    [1.42, 21220, 169.30, 41089, 22.37],
    [1.35,  7394,  39.53, 14257, 28.13],
    [1.39, 11586,  40.41, 22661, 17.55],
    [1.16, 26609, 102.96, 52509, 21.92],
    [1.27,  7801,  37.02, 14903, 19.52],
    [1.16, 11587,  45.74, 25587, 23.99],
    [1.25,  9475,  40.07, 16821, 21.76]
]

X = np.array(data)
n, p = X.shape # n - кількість об'єктів (10), p - кількість ознак (5)
print(f"Розмірність початкової матриці X: {n} об'єктів, {p} ознак.")
```

### 1. Нормалізація початкових даних та розрахунок кореляційної матриці R

Згідно з **Зауваженням 7.2 та 7.3**, щоб уникнути впливу різних одиниць виміру, перейдемо до допоміжних безрозмірних (нормалізованих) ознак за правилом:
$$x^*_{ji} = \frac{x_{ji} - \bar{x}_j}{\sqrt{\hat{\sigma}_j^2}}$$
де $\bar{x}_j$ — середнє значення $j$-ї ознаки, $\hat{\sigma}_j^2 = \frac{1}{n} \sum_{i=1}^n (x_{ji} - \bar{x}_j)^2$ — вибіркова дисперсія.

Після цього розрахуємо кореляційну матрицю $R$. Для нормалізованих даних матриця коваріацій збігається з кореляційною:
$$R = \frac{1}{n} (X^*)^T X^*$$

```python
# Крок 1.1: Нормалізація
# Обчислюємо середнє для кожного стовпця (ознаки)
X_mean = np.mean(X, axis=0)

# Обчислюємо дисперсію для кожного стовпця (ділимо на n, тому ddof=0)
X_var = np.var(X, axis=0, ddof=0)
X_std = np.sqrt(X_var)

# Нормалізована матриця
X_norm = (X - X_mean) / X_std

print("Нормалізована матриця X*:")
print(np.round(X_norm, 4))

# Крок 1.2: Розрахунок кореляційної матриці R
R = (X_norm.T @ X_norm) / n

print("\nКореляційна матриця R:")
df_R = pd.DataFrame(R,
                    columns=['Озн 7', 'Озн 11', 'Озн 12', 'Озн 13', 'Озн 17'],
                    index=['Озн 7', 'Озн 11', 'Озн 12', 'Озн 13', 'Озн 17'])
print(df_R.round(4))
```

### Перевірка значущості кореляційної матриці (Критерій Уїлкса)

Перевірку значущості кореляційної матриці здійснюємо за допомогою критерію Уїлкса.
Спочатку розраховуємо статистику:
$$\chi^2_{\text{ст}} = -\left(n - \frac{1}{6}(2p + 5)\right) \ln |R|$$
де $|R|$ — визначник кореляційної матриці.
Далі розраховуємо критичне значення $\chi^2_{\text{кр}}$ для рівня значущості $\alpha = 0.05$ та числа ступенів свободи $\nu = 0.5 \cdot (p - 1) \cdot p$.
Якщо $\chi^2_{\text{ст}} > \chi^2_{\text{кр}}$, то кореляційна матриця значима і застосування МГК має сенс.

```python
# Розрахунок визначника матриці R
det_R = np.linalg.det(R)

# Розрахунок статистики Уїлкса
chi2_st = - (n - (1/6) * (2*p + 5)) * np.log(det_R)

# Розрахунок ступенів свободи
df_wilks = 0.5 * (p - 1) * p

# Рівень значущості
alpha = 0.05

# Знаходження табличного (критичного) значення
chi2_cr = chi2.ppf(1 - alpha, df_wilks)

print(f"Визначник матриці |R| = {det_R:.6f}")
print(f"Статистика Уїлкса X^2_ст = {chi2_st:.4f}")
print(f"Критичне значення X^2_кр = {chi2_cr:.4f}")

if chi2_st > chi2_cr:
    print("Висновок: X^2_ст > X^2_кр. Кореляційна матриця ЗНАЧИМА з ймовірністю 95%. Можна застосовувати МГК.")
else:
    print("Висновок: X^2_ст <= X^2_кр. Кореляційна матриця НЕ ЗНАЧИМА. Застосування МГК не має сенсу.")
```

### 2. Знаходження власних чисел і векторів. Розрахунок матриці навантажень $A$

Для знаходження головних компонент розв'язуємо задачу на власні числа та власні вектори для матриці $R$. Власні числа $\lambda_i$ сортуються за спаданням. Відповідні їм власні вектори утворюють рядки матриці $L$.

Матриця навантажень $A$ визначається за формулою (розділ 7.3):
$$A = L^T \Lambda^{\frac{1}{2}}$$
де $\Lambda^{\frac{1}{2}}$ — діагональна матрица, на головній діагоналі якої розташовані корені з власних чисел $\sqrt{\lambda_j}$.

```python
# Знаходження власних чисел та векторів
eigenvalues, eigenvectors = np.linalg.eig(R)

# Сортування власних чисел за спаданням та відповідне переставлення векторів
idx_sorted = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[idx_sorted]
eigenvectors = eigenvectors[:, idx_sorted]

# Матриця L: її рядки - це власні вектори (тому транспонуємо матрицю, яку повернув eig)
L = eigenvectors.T

print("Власні числа (lambda):")
print(np.round(eigenvalues, 4))

print("\nМатриця власних векторів L (рядки - вектори, стовпці - ознаки):")
# Виводимо L як гарну таблицю.
# ВАЖЛИВО: для 9 варіанту зміни назви стовпців на ['Озн 5', 'Озн 6', 'Озн 10', 'Озн 15', 'Озн 17']
df_L = pd.DataFrame(L,
                    columns=['Озн 7', 'Озн 11', 'Озн 12', 'Озн 13', 'Озн 17'],
                    index=[f'Вектор {i+1} (L{i+1})' for i in range(p)])
print(df_L.round(4))

# Розрахунок матриці навантажень A
# Використовуємо np.maximum щоб уникнути помилок з мікроскопічними від'ємними числами (через похибку ПК)
Lambda_half = np.diag(np.sqrt(np.maximum(eigenvalues, 0)))
A = L.T @ Lambda_half

print("\nМатриця навантажень A:")
# ВАЖЛИВО: для 9 варіанту зміни індекси на ['Озн 5', 'Озн 6', 'Озн 10', 'Озн 15', 'Озн 17']
df_A = pd.DataFrame(A,
                    columns=[f'Z{i+1}' for i in range(p)],
                    index=['Озн 7', 'Озн 11', 'Озн 12', 'Озн 13', 'Озн 17'])
print(df_A.round(4))
```

### 3. Аналіз матриці A та інтерпретація головних компонент

Згідно з **Наслідком 7.2**, **Критерій інформативності** $m$ головних компонент розраховується за формулою (7.8):
$$I_m (Z(X)) = \frac{\lambda_1 + \dots + \lambda_m}{p}$$
Він показує, яку частку загальної дисперсії пояснюють відібрані компоненти.

Для інтерпретації кожної компоненти відбираються ознаки з найбільшими навантаженнями. Важливість навантажень підтверджується **Коефіцієнтом інформативності** $K_j$:
$$K_j = \frac{\sum_i a_{ij}^2}{\lambda_j}$$
де чисельник обчислюється за значеннями $a_{ij}$, що відібрані для пояснення $j$-ї головної компоненти. Якщо $K_j > 0.75$, набір пояснюючих ознак вважається задовільним.

```python
# 1. Розрахунок Критерію інформативності I_m (Формула 7.8)
I_m = np.cumsum(eigenvalues) / p

print("Критерій інформативності (накопичена частка I_m):")
for i, I in enumerate(I_m):
    # Окрема частка поточної компоненти = lambda_i / p
    current_share = eigenvalues[i] / p
    print(f"z{i+1}: додає {current_share*100:.2f}% -> I_{i+1} = {I*100:.2f}%")

print("\n--- Аналіз навантажень та Коефіцієнт інформативності K_j ---")
# ВАЖЛИВО: Для 9 варіанту замінити масив нижче на ['5', '6', '10', '15', '17']
orig_feature_names = ['7', '11', '12', '13', '17']

# Аналізуємо зв'язок ознак з першими двома компонентами
for j in range(2):
    loadings = np.abs(A[:, j])

    # Вибираємо ознаки, де модуль навантаження > 0.7 (сильний зв'язок)
    significant_features = np.where(loadings > 0.7)[0]

    print(f"\nГоловна компонента z{j+1} пояснюється ознаками:")

    sum_a_sq = 0 # Сума квадратів відібраних навантажень
    for f_idx in significant_features:
        val = A[f_idx, j]
        sum_a_sq += val**2
        print(f" - Ознака {orig_feature_names[f_idx]} (навантаження a_{f_idx+1},{j+1} = {val:.4f})")

    # 2. Розрахунок Коефіцієнта інформативності K_j
    K_j = sum_a_sq / eigenvalues[j]
    print(f"Коефіцієнт інформативності K_{j+1} = {sum_a_sq:.4f} / {eigenvalues[j]:.4f} = {K_j:.4f}")

    # Перевірка умови з підручника (більше 0.75)
    if K_j > 0.75:
        print("Висновок: K_j > 0.75. Набір пояснюючих ознак є ЗАДОВІЛЬНИМ. Інтерпретація коректна.")
    else:
        print("Висновок: K_j <= 0.75. Набір ознак недостатній (слід додати ознаки з меншими навантаженнями).")
```

### 4. Визначення кількості головних компонент за допомогою тесту Лоулі і Максвелла

У випадку використання кореляційної матриці тест для визначення достатності виділених $r_1$ компонент має вигляд:
$$\chi^2_{\text{ст}} = - (n - 1) \left[ \sum_{j=r_1+1}^p \ln \lambda_j - (p - r_1) \ln \frac{\sum_{j=r_1+1}^p \lambda_j}{p - r_1} \right]$$
Ступені свободи: $\nu = 0.5 (p - r_1)(p - r_1 + 1) - 1$.
Перевірятимемо достатність кількості компонент послідовно для $r_1 = 1, 2, ...$ поки не отримаємо $\chi^2_{\text{ст}} < \chi^2_{\text{кр}}$.

```python
print("Тест Лоулі і Максвелла для кореляційної матриці:")

optimal_components = p

for r1 in range(1, p):
    # Беремо власні числа від r1+1 до p (індекси від r1 до p-1)
    lambda_tail = eigenvalues[r1:]

    sum_ln = np.sum(np.log(lambda_tail))
    ln_sum_avg = np.log(np.sum(lambda_tail) / (p - r1))

    # Статистика тесту
    chi2_lm_st = - (n - 1) * (sum_ln - (p - r1) * ln_sum_avg)

    # Ступені свободи
    df_lm = 0.5 * (p - r1) * (p - r1 + 1) - 1

    # Якщо df <= 0 (може бути на останніх кроках), тест не застосовний
    if df_lm <= 0:
        break

    chi2_lm_cr = chi2.ppf(1 - alpha, df_lm)

    print(f"\nПеревірка достатності r1 = {r1} компонент:")
    print(f"X^2_ст = {chi2_lm_st:.4f}")
    print(f"X^2_кр = {chi2_lm_cr:.4f} (df = {df_lm})")

    if chi2_lm_st > chi2_lm_cr:
        print("Висновок: X^2_ст > X^2_кр. Потрібно додати ще компоненти.")
    else:
        print("Висновок: X^2_ст <= X^2_кр. Виділеної кількості компонент ДОСТАТНЬО.")
        optimal_components = r1
        break

print(f"\nОтже, згідно з тестом Лоулі і Максвелла, достатньо залишити {optimal_components} головних компонент.")
```

### 5. Розрахунок матриці головних компонент Z

Оскільки ми попередньо перейшли до нормалізованих змінних $X^*$, матриця головних компонент (значення компонент для кожного об'єкта) розраховується як лінійна комбінація (див. приклад на скріншотах):
$$Z = X^* L^T$$
*(Оскільки в теорії об'єкти розташовані у вигляді стовпців, використовується $Z = LX$. В нашому датасеті об'єкти — це рядки матриці $X^*$, тому ми виконуємо матричне множення $X^* \cdot L^T$)*.

```python
# Розрахунок матриці головних компонент Z
# Розмірність X_norm: (10, 5), розмірність L.T: (5, 5) -> результат (10, 5)
Z = X_norm @ L.T

print(f"Матриця головних компонент Z (розмірність {Z.shape}):")
df_Z = pd.DataFrame(Z,
                    columns=[f'z{i+1}' for i in range(p)],
                    index=[f'Об\'єкт {i+1}' for i in range(n)])
print(df_Z.round(4))

# Якщо ми вирішили залишити лише r1 оптимальних компонент (наприклад, 2),
# то результуюча матриця для подальшого аналізу складатиметься лише з перших r1 стовпців Z.
Z_optimal = Z[:, :optimal_components]
print(f"\nОптимальна матриця Z (лише перші {optimal_components} компоненти):")
df_Z_opt = pd.DataFrame(Z_optimal,
                        columns=[f'z{i+1}' for i in range(optimal_components)],
                        index=[f'Об\'єкт {i+1}' for i in range(n)])
print(df_Z_opt.round(4))
```

Порада до захисту роботи: Викладач може запитати, чому при розрахунку матриці $Z$ (останній крок) використано $X^* L^T$, а не $L X$, як у прикладі з підручника. Відповідь полягає у розмірності даних: у прикладі з книжки вектор-стовпець описує об'єкт, а у нас об'єкти — це рядки матриці pandas/numpy (стандартне подання в аналізі даних). Тому, за правилами лінійної алгебри, щоб отримати лінійну комбінацію ознак у рядках, потрібно домножити матрицю даних справа на транспоновану матрицю векторів.
