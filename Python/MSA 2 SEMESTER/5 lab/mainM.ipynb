# Метод головних компонент (Завдання 14, Варіант 9)
**Мета роботи:** Провести аналіз ознак методом головних компонент (МГК) за математичними формулами (без використання високорівневих пакетів машинного навчання).

### Ініціалізація даних та імпорт бібліотек
Задамо початкову матрицю $X$, де рядки — це об'єкти (1-10), а стовпці — показники підприємств (ознаки 5, 6, 10, 15, 17 з таблиці А.6).

```python
import numpy as np
import pandas as pd
from scipy.stats import chi2 # Використовуємо лише для отримання табличних критичних значень

# Початкові дані (Варіант 9: ознаки 5, 6, 10, 15, 17)
# Рядки - об'єкти (1-10), стовпці - ознаки
data = [
    [0.78, 0.40, 1.45, 166.32, 17.72],
    [0.75, 0.26, 1.30,  92.88, 18.39],
    [0.68, 0.40, 1.37, 158.04, 26.46],
    [0.70, 0.50, 1.65,  93.96, 22.37],
    [0.62, 0.40, 1.91, 173.88, 28.13],
    [0.76, 0.19, 1.68, 162.30, 17.55],
    [0.73, 0.25, 1.94,  88.56, 21.92],
    [0.71, 0.44, 1.89, 101.16, 19.52],
    [0.69, 0.17, 1.94, 166.32, 23.99],
    [0.73, 0.39, 2.06, 140.76, 21.76]
]

X = np.array(data)
n, p = X.shape # n - кількість об'єктів (10), p - кількість ознак (5)
print(f"Розмірність початкової матриці X: {n} об'єктів, {p} ознак.")
```

### 1. Нормалізація початкових даних та розрахунок кореляційної матриці R

Згідно з теорією, переходимо до нормованих ознак:
$$x^*_{ji} = \frac{x_{ji} - \bar{x}_j}{\sqrt{\hat{\sigma}_j^2}}$$
Кореляційна матриця розраховується як:
$$R = \frac{1}{n} (X^*)^T X^*$$

```python
# Крок 1.1: Нормалізація
X_mean = np.mean(X, axis=0)
X_var = np.var(X, axis=0, ddof=0)
X_std = np.sqrt(X_var)

# Нормалізована матриця
X_norm = (X - X_mean) / X_std

print("Нормалізована матриця X* (перші 3 рядки):")
print(np.round(X_norm[:3], 4))

# Крок 1.2: Розрахунок кореляційної матриці R
R = (X_norm.T @ X_norm) / n

print("\nКореляційна матриця R:")
df_R = pd.DataFrame(R,
                    columns=['Озн 5', 'Озн 6', 'Озн 10', 'Озн 15', 'Озн 17'],
                    index=['Озн 5', 'Озн 6', 'Озн 10', 'Озн 15', 'Озн 17'])
print(df_R.round(4))
```

### Перевірка значущості кореляційної матриці (Критерій Уїлкса)

Перевірка за формулою:
$$\chi^2_{\text{ст}} = -\left(n - \frac{1}{6}(2p + 5)\right) \ln |R|$$

```python
det_R = np.linalg.det(R)
chi2_st = - (n - (1/6) * (2*p + 5)) * np.log(det_R)

# Ступені свободи та критичне значення
df_wilks = 0.5 * (p - 1) * p
alpha = 0.05
chi2_cr = chi2.ppf(1 - alpha, df_wilks)

print(f"Визначник матриці |R| = {det_R:.6f}")
print(f"Статистика Уїлкса X^2_ст = {chi2_st:.4f}")
print(f"Критичне значення X^2_кр = {chi2_cr:.4f}")

if chi2_st > chi2_cr:
    print("Висновок: X^2_ст > X^2_кр. Кореляційна матриця ЗНАЧИМА з ймовірністю 95%. Можна застосовувати МГК.")
else:
    print("Висновок: X^2_ст <= X^2_кр. Кореляційна матриця НЕ ЗНАЧИМА. Зв'язки між ознаками надто слабкі.")
```

### 2. Знаходження власних чисел і векторів. Розрахунок матриці навантажень $A$

Матриця навантажень $A$ визначається за формулою:
$$A = L^T \Lambda^{\frac{1}{2}}$$

```python
# Знаходження власних чисел та векторів
eigenvalues, eigenvectors = np.linalg.eig(R)

# Сортування за спаданням
idx_sorted = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[idx_sorted]
eigenvectors = eigenvectors[:, idx_sorted]

# Матриця L
L = eigenvectors.T

print("Власні числа (lambda):")
print(np.round(eigenvalues, 4))

# Розрахунок матриці навантажень A
Lambda_half = np.diag(np.sqrt(np.maximum(eigenvalues, 0))) # maximum для уникнення проблем з мікроскопічними від'ємними числами через похибки
A = L.T @ Lambda_half

print("\nМатриця навантажень A:")
df_A = pd.DataFrame(A,
                    columns=[f'Z{i+1}' for i in range(p)],
                    index=['Озн 5', 'Озн 6', 'Озн 10', 'Озн 15', 'Озн 17'])
print(df_A.round(4))
```

### 3. Аналіз матриці A та інтерпретація головних компонент

Коефіцієнт інформативності: $I_j = \frac{\lambda_j}{p}$

```python
inform_coeffs = eigenvalues / p
cumulative_inform = np.cumsum(inform_coeffs)

print("Інформативність кожної головної компоненти:")
for i, inf in enumerate(inform_coeffs):
    print(f"z{i+1}: {inf*100:.2f}% (накопичена: {cumulative_inform[i]*100:.2f}%)")

print("\n--- Аналіз навантажень ---")
# Аналізуємо зв'язок ознак з першими двома компонентами
for j in range(2):
    loadings = np.abs(A[:, j])
    significant_features = np.where(loadings > 0.65)[0] # Поріг 0.65 для виявлення значущих зв'язків

    print(f"\nГоловна компонента z{j+1} сильно корелює з ознаками:")
    orig_feature_names = ['5', '6', '10', '15', '17']
    for f_idx in significant_features:
        val = A[f_idx, j]
        print(f" - Ознака {orig_feature_names[f_idx]} (навантаження: {val:.4f})")

print("\nІнтерпретація:")
print("Виходячи з навантажень, перші компоненти групують показники за прихованими чинниками діяльності підприємств. Ті ознаки, які мають велике за модулем навантаження на одну компоненту, пояснюються одним спільним латентним фактором.")
```

### 4. Визначення кількості головних компонент за допомогою тесту Лоулі і Максвелла

Тест для визначення достатності виділених $r_1$ компонент:
$$\chi^2_{\text{ст}} = - (n - 1) \left[ \sum_{j=r_1+1}^p \ln \lambda_j - (p - r_1) \ln \frac{\sum_{j=r_1+1}^p \lambda_j}{p - r_1} \right]$$

```python
print("Тест Лоулі і Максвелла для кореляційної матриці:")

optimal_components = p

for r1 in range(1, p):
    lambda_tail = eigenvalues[r1:]

    # Захист від від'ємних власних чисел (через обчислювальні похибки матриці)
    if np.any(lambda_tail <= 0):
        print(f"\nРешта власних чисел близькі до нуля. Зупиняємось на {r1} компонентах.")
        optimal_components = r1
        break

    sum_ln = np.sum(np.log(lambda_tail))
    ln_sum_avg = np.log(np.sum(lambda_tail) / (p - r1))

    chi2_lm_st = - (n - 1) * (sum_ln - (p - r1) * ln_sum_avg)
    df_lm = 0.5 * (p - r1) * (p - r1 + 1) - 1

    if df_lm <= 0:
        break

    chi2_lm_cr = chi2.ppf(1 - alpha, df_lm)

    print(f"\nПеревірка достатності r1 = {r1} компонент:")
    print(f"X^2_ст = {chi2_lm_st:.4f}, X^2_кр = {chi2_lm_cr:.4f} (df = {df_lm})")

    if chi2_lm_st > chi2_lm_cr:
        print("Висновок: X^2_ст > X^2_кр. Потрібно додати ще компоненти.")
    else:
        print("Висновок: X^2_ст <= X^2_кр. Виділеної кількості компонент ДОСТАТНЬО.")
        optimal_components = r1
        break

print(f"\nОтже, достатньо залишити {optimal_components} головних компонент.")
```

### 5. Розрахунок матриці головних компонент Z

Матриця значень головних компонент розраховується як:
$$Z = X^* L^T$$

```python
# Розрахунок матриці головних компонент Z
Z = X_norm @ L.T

print(f"Матриця головних компонент Z (всі {p} компонент):")
df_Z = pd.DataFrame(Z,
                    columns=[f'z{i+1}' for i in range(p)],
                    index=[f'Об\'єкт {i+1}' for i in range(n)])
print(df_Z.round(4))

# Зріз оптимальної кількості компонент
Z_optimal = Z[:, :optimal_components]
print(f"\nОптимальна матриця Z (лише перші {optimal_components} компоненти):")
df_Z_opt = pd.DataFrame(Z_optimal,
                        columns=[f'z{i+1}' for i in range(optimal_components)],
                        index=[f'Об\'єкт {i+1}' for i in range(n)])
print(df_Z_opt.round(4))
```

Порада до захисту роботи: Викладач може запитати, чому при розрахунку матриці $Z$ (останній крок) використано $X^* L^T$, а не $L X$, як у прикладі з підручника. Відповідь полягає у розмірності даних: у прикладі з книжки вектор-стовпець описує об'єкт, а у нас об'єкти — це рядки матриці pandas/numpy (стандартне подання в аналізі даних). Тому, за правилами лінійної алгебри, щоб отримати лінійну комбінацію ознак у рядках, потрібно домножити матрицю даних справа на транспоновану матрицю векторів.
