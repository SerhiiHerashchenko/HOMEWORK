# Метод головних компонент (Завдання 14, Варіант 9)
**Мета роботи:** Провести аналіз ознак методом головних компонент (МГК) за математичними формулами (без використання високорівневих пакетів машинного навчання).

### Ініціалізація даних та імпорт бібліотек
Задамо початкову матрицю $X$, де рядки — це об'єкти (1-10), а стовпці — показники підприємств (ознаки 5, 6, 10, 15, 17 з таблиці А.6).

```python
import numpy as np
import pandas as pd
from scipy.stats import chi2 # Використовуємо лише для отримання табличних критичних значень

# Початкові дані (Варіант 9: ознаки 5, 6, 10, 15, 17)
# Рядки - об'єкти (1-10), стовпці - ознаки
data = [
    [0.78, 0.40, 1.45, 166.32, 17.72],
    [0.75, 0.26, 1.30,  92.88, 18.39],
    [0.68, 0.40, 1.37, 158.04, 26.46],
    [0.70, 0.50, 1.65,  93.96, 22.37],
    [0.62, 0.40, 1.91, 173.88, 28.13],
    [0.76, 0.19, 1.68, 162.30, 17.55],
    [0.73, 0.25, 1.94,  88.56, 21.92],
    [0.71, 0.44, 1.89, 101.16, 19.52],
    [0.69, 0.17, 1.94, 166.32, 23.99],
    [0.73, 0.39, 2.06, 140.76, 21.76]
]

X = np.array(data)
n, p = X.shape # n - кількість об'єктів (10), p - кількість ознак (5)
print(f"Розмірність початкової матриці X: {n} об'єктів, {p} ознак.")
```

### 1. Нормалізація початкових даних та розрахунок кореляційної матриці R

Згідно з теорією, переходимо до нормованих ознак:
$$x^*_{ji} = \frac{x_{ji} - \bar{x}_j}{\sqrt{\hat{\sigma}_j^2}}$$
Кореляційна матриця розраховується як:
$$R = \frac{1}{n} (X^*)^T X^*$$

```python
# Крок 1.1: Нормалізація
X_mean = np.mean(X, axis=0)
X_var = np.var(X, axis=0, ddof=0)
X_std = np.sqrt(X_var)

# Нормалізована матриця
X_norm = (X - X_mean) / X_std

print("Нормалізована матриця X*:")
print(np.round(X_norm, 4))

# Крок 1.2: Розрахунок кореляційної матриці R
R = (X_norm.T @ X_norm) / n

print("\nКореляційна матриця R:")
df_R = pd.DataFrame(R,
                    columns=['Озн 5', 'Озн 6', 'Озн 10', 'Озн 15', 'Озн 17'],
                    index=['Озн 5', 'Озн 6', 'Озн 10', 'Озн 15', 'Озн 17'])
print(df_R.round(4))
```

### Перевірка значущості кореляційної матриці (Критерій Уїлкса)

Перевірка за формулою:
$$\chi^2_{\text{ст}} = -\left(n - \frac{1}{6}(2p + 5)\right) \ln |R|$$

```python
det_R = np.linalg.det(R)
chi2_st = - (n - (1/6) * (2*p + 5)) * np.log(det_R)

# Ступені свободи та критичне значення
df_wilks = 0.5 * (p - 1) * p
alpha = 0.05
chi2_cr = chi2.ppf(1 - alpha, df_wilks)

print(f"Визначник матриці |R| = {det_R:.6f}")
print(f"Статистика Уїлкса X^2_ст = {chi2_st:.4f}")
print(f"Критичне значення X^2_кр = {chi2_cr:.4f}")

if chi2_st > chi2_cr:
    print("Висновок: X^2_ст > X^2_кр. Кореляційна матриця ЗНАЧИМА з ймовірністю 95%. Можна застосовувати МГК.")
else:
    print("Висновок: X^2_ст <= X^2_кр. Кореляційна матриця НЕ ЗНАЧИМА. Зв'язки між ознаками надто слабкі.")
```

### 2. Знаходження власних чисел і векторів. Розрахунок матриці навантажень $A$

Матриця навантажень $A$ визначається за формулою:
$$A = L^T \Lambda^{\frac{1}{2}}$$

```python
# Знаходження власних чисел та векторів
eigenvalues, eigenvectors = np.linalg.eig(R)

# Сортування власних чисел за спаданням та відповідне переставлення векторів
idx_sorted = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[idx_sorted]
eigenvectors = eigenvectors[:, idx_sorted]

# Матриця L: її рядки - це власні вектори (тому транспонуємо матрицю, яку повернув eig)
L = eigenvectors.T

print("Власні числа (lambda):")
print(np.round(eigenvalues, 4))

print("\nМатриця власних векторів L (рядки - вектори, стовпці - ознаки):")
# Виводимо L як гарну таблицю.
# ВАЖЛИВО: для 9 варіанту зміни назви стовпців на ['Озн 5', 'Озн 6', 'Озн 10', 'Озн 15', 'Озн 17']
df_L = pd.DataFrame(L,
                    columns=['Озн 5', 'Озн 6', 'Озн 10', 'Озн 15', 'Озн 17'],
                    index=[f'Вектор {i+1} (L{i+1})' for i in range(p)])
print(df_L.round(4))

# Розрахунок матриці навантажень A
# Використовуємо np.maximum щоб уникнути помилок з мікроскопічними від'ємними числами (через похибку ПК)
Lambda_half = np.diag(np.sqrt(np.maximum(eigenvalues, 0)))
A = L.T @ Lambda_half

print("\nМатриця навантажень A:")
# ВАЖЛИВО: для 9 варіанту зміни індекси на ['Озн 5', 'Озн 6', 'Озн 10', 'Озн 15', 'Озн 17']
df_A = pd.DataFrame(A,
                    columns=[f'Z{i+1}' for i in range(p)],
                    index=['Озн 5', 'Озн 6', 'Озн 10', 'Озн 15', 'Озн 17'])
print(df_A.round(4))
```

### 3. Аналіз матриці A та інтерпретація головних компонент

Згідно з **Наслідком 7.2**, **Критерій інформативності** $m$ головних компонент розраховується за формулою (7.8):
$$I_m (Z(X)) = \frac{\lambda_1 + \dots + \lambda_m}{p}$$
Він показує, яку частку загальної дисперсії пояснюють відібрані компоненти.

Для інтерпретації кожної компоненти відбираються ознаки з найбільшими навантаженнями. Важливість навантажень підтверджується **Коефіцієнтом інформативності** $K_j$:
$$K_j = \frac{\sum_i a_{ij}^2}{\lambda_j}$$
де чисельник обчислюється за значеннями $a_{ij}$, що відібрані для пояснення $j$-ї головної компоненти. Якщо $K_j > 0.75$, набір пояснюючих ознак вважається задовільним.

```python
# 1. Розрахунок Критерію інформативності I_m (Формула 7.8)
I_m = np.cumsum(eigenvalues) / p

print("Критерій інформативності (накопичена частка I_m):")
for i, I in enumerate(I_m):
    # Окрема частка поточної компоненти = lambda_i / p
    current_share = eigenvalues[i] / p
    print(f"z{i+1}: додає {current_share*100:.2f}% -> I_{i+1} = {I*100:.2f}%")

print("\n--- Аналіз навантажень та Коефіцієнт інформативності K_j ---")
# Назви ознак для 9 варіанту
orig_feature_names = ['5', '6', '10', '15', '17']

# Аналізуємо зв'язок ознак з першими двома компонентами
for j in range(3):
    print(f"\n>>> Формуємо інтерпретацію для Головної компоненти z{j+1}:")

    # Отримуємо модулі навантажень для поточної компоненти
    loadings_abs = np.abs(A[:, j])

    # Сортуємо індекси ознак за спаданням модуля навантаження
    # щоб додавати спочатку найвагоміші ознаки
    sorted_indices = np.argsort(loadings_abs)[::-1]

    sum_a_sq = 0 # Сума квадратів відібраних навантажень
    selected_features = [] # Список відібраних ознак

    print("Послідовно додаємо ознаки з найбільшими навантаженнями:")
    for idx in sorted_indices:
        val = A[idx, j]
        selected_features.append(orig_feature_names[idx])
        sum_a_sq += val**2

        # 2. Розрахунок Коефіцієнта інформативності K_j
        K_j = sum_a_sq / eigenvalues[j]

        print(f" + Додано Ознаку {orig_feature_names[idx]} (навантаження a_{idx+1},{j+1} = {val:.4f})")
        print(f"   Поточний коефіцієнт K_{j+1} = {sum_a_sq:.4f} / {eigenvalues[j]:.4f} = {K_j:.4f}")

        # Перевірка умови з підручника
        if K_j > 0.75:
            print(f" => ВИСНОВОК: K_{j+1} > 0.75. Набір ознак {selected_features} є ЗАДОВІЛЬНИМ. Відбір завершено.")
            break
        else:
            print(f" -> K_{j+1} <= 0.75. Набір недостатній. Додаємо наступну ознаку...")
```

### 4. Визначення кількості головних компонент

**1. Тест Лоулі і Максвелла.** У випадку використання кореляційної матриці тест для визначення достатності виділених $r_1$ компонент має вигляд:
$$\chi^2_{\text{ст}} = - (n - 1) \left[ \sum_{j=r_1+1}^p \ln \lambda_j - (p - r_1) \ln \frac{\sum_{j=r_1+1}^p \lambda_j}{p - r_1} \right]$$
Ступені свободи: $\nu = 0.5 (p - r_1)(p - r_1 + 1) - 1$.
Алгоритм додає по одній компоненті ($r_1 = 1, 2, ...$) і перевіряє умову $\chi^2_{\text{ст}} \le \chi^2_{\text{кр}}$. Якщо умова не виконується — додаємо наступну компоненту.

**2. Практичне правило.** Як зазначено в теорії, якщо тест вимагає занадто багато компонент, на практиці використовують умову: якщо критерій інформативності $I_m > 0.80$, то взятої кількості головних компонент достатньо.

```python
print("--- Тест Лоулі і Максвелла ---")

optimal_components = None
alpha = 0.05

# Ми можемо тестувати r1 лише до p-2, бо далі ступені свободи (df) стають <= 0
for r1 in range(1, p - 1):
    lambda_tail = eigenvalues[r1:]

    # Розрахунок формули зі скріншота
    sum_ln = np.sum(np.log(lambda_tail))
    ln_sum_avg = np.log(np.sum(lambda_tail) / (p - r1))

    chi2_lm_st = - (n - 1) * (sum_ln - (p - r1) * ln_sum_avg)
    df_lm = int(0.5 * (p - r1) * (p - r1 + 1) - 1)

    chi2_lm_cr = chi2.ppf(1 - alpha, df_lm)

    print(f"\nКрок {r1}: Перевіряємо достатність r1 = {r1} компонент:")
    print(f"X^2_ст = {chi2_lm_st:.4f}")
    print(f"X^2_кр = {chi2_lm_cr:.4f} (df = {df_lm})")

    if chi2_lm_st > chi2_lm_cr:
        print("Висновок: X^2_ст > X^2_кр. Недостатньо! Необхідно ввести додатково інші головні компоненти.")
    else:
        print("Висновок: X^2_ст <= X^2_кр. Виділеної кількості компонент ДОСТАТНЬО.")
        optimal_components = r1
        break

# Якщо цикл закінчився, а оптимальну кількість так і не знайшли
if optimal_components is None:
    print("\nУВАГА: Тест Лоулі-Максвелла вимагає подальшого додавання компонент, але ступені свободи вичерпано.")
    print("Переходимо до практичного правила з теорії (I_m > 0.80):")

    I_m = np.cumsum(eigenvalues) / p
    for m, I in enumerate(I_m):
        r1_practical = m + 1
        print(f"Перевіряємо m = {r1_practical}: Критерій інформативності I_{r1_practical} = {I:.4f}")
        if I > 0.80:
            print(f"Висновок: Оскільки {I:.4f} > 0.80, кількості компонент m = {r1_practical} достатньо.")
            optimal_components = r1_practical
            break

print(f"\n>>> ПІДСУМОК: Для побудови матриці Z залишаємо {optimal_components} головних компонент.")
```

### 5. Розрахунок матриці головних компонент Z

Матриця значень головних компонент розраховується як:
$$Z = X^* L^T$$

```python
# Розрахунок матриці головних компонент Z
Z = X_norm @ L.T

print(f"Матриця головних компонент Z (всі {p} компонент):")
df_Z = pd.DataFrame(Z,
                    columns=[f'z{i+1}' for i in range(p)],
                    index=[f'Об\'єкт {i+1}' for i in range(n)])
print(df_Z.round(4))

# Зріз оптимальної кількості компонент
Z_optimal = Z[:, :optimal_components]
print(f"\nОптимальна матриця Z (лише перші {optimal_components} компоненти):")
df_Z_opt = pd.DataFrame(Z_optimal,
                        columns=[f'z{i+1}' for i in range(optimal_components)],
                        index=[f'Об\'єкт {i+1}' for i in range(n)])
print(df_Z_opt.round(4))
```

Порада до захисту роботи: Викладач може запитати, чому при розрахунку матриці $Z$ (останній крок) використано $X^* L^T$, а не $L X$, як у прикладі з підручника. Відповідь полягає у розмірності даних: у прикладі з книжки вектор-стовпець описує об'єкт, а у нас об'єкти — це рядки матриці pandas/numpy (стандартне подання в аналізі даних). Тому, за правилами лінійної алгебри, щоб отримати лінійну комбінацію ознак у рядках, потрібно домножити матрицю даних справа на транспоновану матрицю векторів.
