# Завдання 11. Кластерний аналіз ітераційними методами

**Варіант 4:** Об'єкти 31-40, Ознаки: 1, 6, 8, 9.

**Важливе зауваження щодо даних:** Оскільки соціально-економічні показники мають різний масштаб (наприклад, ознака 1 вимірюється в десятках тисяч, а ознака 8 — у десятках), використання евклідової метрики на «сирих» даних призведе до того, що ознаки з великими числами повністю нівелюють інші. Тому, для коректної математичної роботи алгоритмів, перед обчисленням відстаней ми **стандартизуємо** дані за формулою $Z = \frac{x - \mu}{\sigma}$. Це не порушує логіку з теорії, але робить результати змістовними.

```python
import numpy as np

# Налаштування формату виводу numpy для зручності читання (3 знаки після коми)
np.set_printoptions(precision=3, suppress=True)

# Вхідні дані для Варіанту 4 (рядки з 31 по 40 з таблиці А.4)
# Ознаки: 1, 6, 8, 9
raw_data = np.array([
    [3524, 14381, 84, 99],   # 31 Нова Зеландія
    [4300, 17755, 75, 99],   # 32 Норвегія
    [2800, 14193, 81, 68],   # 33 ОАЕ
    [43900, 3128, 49, 76],   # 34 ПАР
    [45000, 6627, 72, 96],   # 35 Південна Корея
    [23100, 1000, 60, 99],   # 36 Північна Корея
    [38600, 4429, 62, 99],   # 37 Польща
    [10500, 9000, 34, 85],   # 38 Португалія
    [149200, 6680, 74, 99],  # 39 Росія
    [18000, 6651, 77, 62]    # 40 Саудівська Аравія
], dtype=float)

names = ["Нова Зеландія", "Норвегія", "ОАЕ", "ПАР", "Південна Корея",
         "Північна Корея", "Польща", "Португалія", "Росія", "Сауд Аравія"]

print("--- ВХІДНІ ДАНІ (перші 3 об'єкти) ---")
print(raw_data[:3])

# Стандартизація даних (Z-score)
def standardize(X):
    means = np.mean(X, axis=0)
    stds = np.std(X, axis=0)
    print("\n--- ПАРАМЕТРИ СТАНДАРТИЗАЦІЇ ---")
    print(f"Середні значення за ознаками (mu): {means}")
    print(f"Середньоквадратичні відхилення (sigma): {stds}")
    return (X - means) / stds

X = standardize(raw_data)

print("\n--- СТАНДАРТИЗОВАНІ ДАНІ (Матриця X) ---")
for i, row in enumerate(X):
    print(f"X_{i+1} ({names[i]:<15}): {row}")

def euclidean_distance(v1, v2):
    return np.sqrt(np.sum((v1 - v2) ** 2))
```

### 1. Класифікація методом $k$-середніх (алгоритм із послідовним перерахунком)

Алгоритм реалізовано **строго за методичкою**:
1. Початкові еталони — перші $k$ об'єктів. Початкові ваги $w = 1$.
2. **Перша ітерація:** розглядаються об'єкти починаючи з $k+1$. Вони приєднуються до найближчого еталона, після чого еталон *одразу* перераховується з урахуванням накопиченої ваги: $\boldsymbol{E}_j = \frac{w_j \boldsymbol{E}_j + \boldsymbol{n}_{k+1}}{w_j + 1}$.
3. **Наступні ітерації:** переглядається *вся* сукупність об'єктів ($n_1 \dots n_m$). Ваги продовжують накопичуватися.
4. Процес зупиняється, коли розбиття на поточній ітерації співпадає з попереднім.
5. **Остаточне розбиття:** формуються класифікаційні центри ваги (вже без історичного накопичення), і об'єкти остаточно розподіляються за мінімальною відстанню до них.

```python
def kmeans_sequential(X, k=2):
    m, n = X.shape

    # 1. Ініціалізація
    E = X[:k].copy()
    w = np.ones(k)

    assignments = np.zeros(m, dtype=int)
    for i in range(k):
        assignments[i] = i

    print(f"Початкові еталони E_1, E_2 (взяті з об'єктів 1 та 2):\n{E}")
    print(f"Початкові ваги w: {w}\n")

    print(f"=== ПЕРША ІТЕРАЦІЯ (Послідовне приєднання) ===")
    for i in range(k, m):
        dists = [euclidean_distance(X[i], E[j]) for j in range(k)]
        c = np.argmin(dists)
        assignments[i] = c

        print(f"Об'єкт {i+1} ({names[i]}):")
        print(f"  Відстані до поточних еталонів: d1={dists[0]:.3f}, d2={dists[1]:.3f}")
        print(f"  -> Приєднуємо до Класу {c+1}")

        # Перерахунок еталона (формула з теорії)
        old_E = E[c].copy()
        E[c] = (w[c] * E[c] + X[i]) / (w[c] + 1)
        w[c] += 1
        print(f"  -> Перераховано еталон E_{c+1}: змістився на {euclidean_distance(old_E, E[c]):.3f}")
        print(f"  -> Нова вага класу {c+1}: w={w[c]}\n")

    print(f"Розбиття після 1-ї ітерації: {assignments + 1}")

    prev_assignments = assignments.copy()
    iteration = 2

    # 3. Наступні ітерації
    while True:
        print(f"\n=== ІТЕРАЦІЯ {iteration} (Повторний перерозподіл) ===")
        current_assignments = np.zeros(m, dtype=int)
        for i in range(m):
            dists = [euclidean_distance(X[i], E[j]) for j in range(k)]
            c = np.argmin(dists)
            current_assignments[i] = c

            E[c] = (w[c] * E[c] + X[i]) / (w[c] + 1)
            w[c] += 1

        print(f"Розбиття: {current_assignments + 1}")
        print(f"Накопичені ваги еталонів: {w}")

        if np.array_equal(current_assignments, prev_assignments):
            print(f"-> Класи не змінилися. Збіжність досягнута.")
            break

        prev_assignments = current_assignments.copy()
        iteration += 1

    # 4. Остаточне розбиття
    print(f"\n=== ОСТАТОЧНЕ РОЗБИТТЯ (за центрами ваги класів) ===")
    final_centroids = np.zeros((k, n))
    for j in range(k):
        cluster_points = X[current_assignments == j]
        if len(cluster_points) > 0:
            final_centroids[j] = np.mean(cluster_points, axis=0)
            print(f"Остаточний центр мас c_{j+1}:\n  {final_centroids[j]}")

    final_assignments = np.zeros(m, dtype=int)
    for i in range(m):
        dists = [euclidean_distance(X[i], final_centroids[j]) for j in range(k)]
        final_assignments[i] = np.argmin(dists)

    return final_assignments

k_classes = kmeans_sequential(X, k=2)

print("\n--- РЕЗУЛЬТАТ k-СЕРЕДНІХ ---")
for i in range(2):
    cluster_names = [names[j] for j in range(len(names)) if k_classes[j] == i]
    print(f"Клас {i+1}: {cluster_names}")
```

### 2. Класифікація методом пошуку згущень (ФОРЕЛ)

Алгоритм реалізовано за такими кроками:
1. Будується матриця відстаней $D$ для визначення $R_H$ (мінімум) та $R_B$ (максимум). Радіус $R$ обирається між цими значеннями.
2. Перший доступний об'єкт стає центром гіперсфери.
3. Шукаються всі об'єкти, що потрапили в сферу радіусом $R$. Розраховується новий центр ваги цих об'єктів.
4. Якщо центр зсунувся — сфера переміщується, і крок 3 повторюється. Якщо ні — формується стійкий клас.
5. Клас виключається з подальшого розгляду, процес повторюється для решти точок.

```python
def forel(X, R_coeff=0.35):
    m, n = X.shape

    # Побудова матриці відстаней
    D = np.zeros((m, m))
    for i in range(m):
        for j in range(m):
            D[i, j] = euclidean_distance(X[i], X[j])

    print("--- МАТРИЦЯ ВІДСТАНЕЙ D (фрагмент 5x5 для об'єктів 1-5) ---")
    print(D[:5, :5])

    # Визначаємо R_H та R_B
    D_masked = np.ma.masked_equal(D, 0.0, copy=False)
    R_H = D_masked.min()
    R_B = D.max()
    R = R_H + R_coeff * (R_B - R_H)

    print(f"\n--- ВИЗНАЧЕННЯ РАДІУСА R ---")
    print(f"Мінімальна відстань R_H = {R_H:.3f}")
    print(f"Максимальна відстань R_B = {R_B:.3f}")
    print(f"Коефіцієнт (дельта) = {R_coeff}")
    print(f"Обраний радіус гіперсфери R = {R:.3f}\n")

    remaining_indices = list(range(m))
    clusters = []
    class_num = 1

    while remaining_indices:
        print(f"=== ФОРМУВАННЯ КЛАСУ {class_num} ===")
        center_idx = remaining_indices[0]
        center = X[center_idx].copy()
        print(f"Початковий центр гіперсфери: об'єкт {center_idx+1} ({names[center_idx]})")

        step = 1
        while True:
            # Шукаємо об'єкти в сфері
            in_sphere = []
            dists_to_center = []
            for idx in remaining_indices:
                dist = euclidean_distance(X[idx], center)
                if dist <= R:
                    in_sphere.append(idx)
                    dists_to_center.append(dist)

            print(f"  Крок {step}: У сферу потрапили об'єкти з індексами {np.array(in_sphere)+1}")
            print(f"           Відстані до центру: {np.round(dists_to_center, 3)}")

            # Новий центр ваги
            new_center = np.mean(X[in_sphere], axis=0)

            if np.allclose(center, new_center):
                print("  -> Центр ваги не змінився. Стійке згущення знайдено.")
                break

            dist_moved = euclidean_distance(center, new_center)
            print(f"  -> Центр зсунувся на {dist_moved:.3f}. Перераховуємо гіперсферу з нового центру.")
            center = new_center
            step += 1

        clusters.append(in_sphere)
        class_num += 1

        # Виключаємо розподілені об'єкти
        for idx in in_sphere:
            remaining_indices.remove(idx)
        print(f"Залишились нерозподілені об'єкти: {np.array(remaining_indices)+1}\n")

    return clusters

forel_clusters_idx = forel(X, R_coeff=0.35)

print("--- РЕЗУЛЬТАТИ ФОРЕЛ ---")
for i, cluster in enumerate(forel_clusters_idx):
    cluster_names = [names[j] for j in cluster]
    print(f"Клас {i+1}: {cluster_names}")
```

**(Примітка для захисту: уважно проаналізуй вивід у попередніх комірках. Зазвичай алгоритми групують ці країни так:)**

1. **Економічно розвинені країни (або країни з високими показниками життя):** До одного класу, як правило, потрапляють такі країни як Норвегія, Нова Зеландія. Вони мають стабільно високі показники за ознаками 6, 8 та 9 (які ймовірно відображають ВВП на душу населення, індекс соціального розвитку, тривалість життя чи подібні метрики).

2. **Країни, що розвиваються, або країни з іншою структурою економіки:** До іншого класу об'єднуються ПАР, Росія, Північна Корея тощо. Вони можуть мати високі значення абсолютних показників (наприклад, ознака 1 — загальна площа або населення), але значно нижчі відносні соціальні показники.

**Порівняння методів:**
* **$k$-середніх** жорстко розбиває простір на задану кількість кластерів ($k=2$). Його мінус у тому, що ми маємо вгадати кількість класів, а початковий розподіл залежить від того, які об'єкти стояли першими у списку.
* **ФОРЕЛ (пошук згущень)** більш гнучкий. Він сам визначає природну кількість кластерів залежно від заданого радіусу сфери. Якщо змінити радіус $R$ у коді, можна побачити утворення мікро-кластерів або виявлення «аномальних» країн (викидів), які не схожі на жодну іншу групу і формують клас з одного об'єкта.
